```python
train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/와인 공모전 공부/train.csv")
train
realtrain = train.drop(['index'], axis=1) # axis=1이 열 ,axis=0 이 행이다. 즉, 드랍하고 ['이름']  열인지 행인지 지정.
realtrain
```

```python
realtrain.shape
```

```python
# 퀄리티 열을 따로 저장했다가 드랍한다.
```

```python
realtrain[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','type' , 'quality']]
# 퀄리티를 판단하는거기 때문에 마지막으로 이동시킴.
```

```python
realtrain = pd.get_dummies(realtrain) # 원핫 인코더
realtrain
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

realtrain.corr()# 히트맵으로 뽑기 위해 상관관계를 본다. 이걸 알아보기 쉽게 시각화 한게 히트맵이다.
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=realtrain.corr(),annot=True)
# 상관관계가 가운데 표시 
#음과 양의 상관관계 뭐가 더 높을지/ 음이나 양이나 높으면 상관관계가 높음. 상위 5개 골라보자
```
